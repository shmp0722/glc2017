```python
correlation_matrix = np.corrcoef(df)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn import datasets
%matplotlib inline
```

```python
# 相関行列のヒートマップを描く
sns.heatmap(correlation_matrix, annot=True, xticklabels=hd, yticklabels=hd)
plt.show()
```
